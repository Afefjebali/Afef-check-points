```python
#Q5: Create a scatter plot to show the relationship between North American sales and European sales.
import plotly.graph_objs as go
import plotly.offline as pyo
trace = go.Scatter(x=df["NA_Sales"], y=df["EU_Sales"])
```

```python
data = [trace]
```

```python
layout = go.Layout(title="Relationship between North American sales and European sales",
                   xaxis=dict(title="NA_Sales"),
                   yaxis=dict(title="EU_Sales"))
```

```python
fig = go.Figure(data=data, layout=layout)
```

```python
pyo.plot(fig)
```

```python
#Q8: Create a pie chart to show the distribution of genres by global sales.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(file_path)
```

```python
Genre=["Genre"]
Global_Sales=["Global_Sales"]
genre_sales = df.groupby("Genre")["Global_Sales"].sum(Global_Sales)
```

```python
plt.pie(genre_sales, labels=genre_sales.index, autopct='%1.1f%%', startangle=140)
plt.title("Distribution of Genres by Global Sales") #The autopct parameter is used to display the percentage of each genre's global sales in the total global sales.
```
